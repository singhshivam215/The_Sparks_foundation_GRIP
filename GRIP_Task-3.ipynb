# Task-3 : Explore Unsupervised Machine Learning

# Import all libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


from sklearn.cluster import KMeans
```

# Reading the Dataset

```python
dataset=pd.read_csv("Iris.csv")
dataset.head(10)
```

```python
dataset.shape
```

```python
dataset.describe()
```

# Data Visualization

```python
sns.pairplot(data=dataset,hue='Species')
```

```python
dataset.info()
```

```python
dataset.drop(['Id'],axis=1,inplace=True)
X=dataset.iloc[:,1:4].values
```

```python
dataset["Species"].unique()
```

# Use elbow method to find optimal number of clusters

```python
wscc=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(X)
    wscc.append(kmeans.inertia_)
plt.plot(range(1,11),wscc)
plt.plot(range(3,4),range(81,82),'ro')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

from this graph ,it can be observed that the elbow position gives us the optimal number of cluster which for this dataset is 3

# Applying K-means

```python
kmeans=KMeans(n_clusters=3,init='k-means++',max_iter=300,n_init=10,random_state=0)
y_kmeans=kmeans.fit_predict(X)
```

# Visualizing cluster

```python
sns.set_style(style='darkgrid')
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=120,c='blue',label="Iris-setosa")
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=120,c='green',label="Iris-versicolor")
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=120,c='maroon',label="Iris-virginica")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200,c='yellow',label='Centroids')
plt.title('SepalLength-SepalWidth plot(Predicted after clustering)')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.legend()
plt.show()
```
